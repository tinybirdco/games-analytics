```python
import pandas as pd

from datetime import datetime, timedelta

import numpy as np
```

```python
def pp(start: datetime, end: datetime, n: int):
    # based on https://stackoverflow.com/questions/50559078/generating-random-dates-within-a-given-range-in-pandas
    start_u = int(start.timestamp())
    end_u = int(end.timestamp())
    return pd.DatetimeIndex((np.random.randint(start_u, end_u, n, dtype=np.int64)).view('M8[s]'))
```

```python
now = datetime.utcnow()
one_year_ago = datetime.utcnow() - timedelta(days=365)

pp(start=one_year_ago, end=now, n=100)
```

## Generating dataset

```python
num_players = 10000
num_teams = num_players // 20

num_players / num_teams
```

### generating random items

```python
import random
```

```python
def generate_random(input_vals: list, n: int, add_num_end: bool = True):
    names = random.choices(input_vals, k=n)
    if add_num_end:
        result = [f'{n}_{i}' for i, n in enumerate(names)]
    else:
        result = names
    return result

generate_random(teams, 10)
```

### game - team

```python
all_teams = generate_random(teams, num_teams)
all_teams
```

```python
team_game_df = pd.DataFrame({'team': all_teams})
team_game_df['game'] = generate_random(game_names, n=num_teams, add_num_end=False)
team_game_df
```

```python
team_game_df.dtypes
```

### people-teams

```python
nicks = generate_random(names_list, n=num_players)
nicks
```

```python
team_game_df.team.sample(num_players, replace=True)
```

```python
people_team_df = pd.DataFrame({'nick': nicks}) 
# multiplying so that people are more than 1 team, playing different games
people_team_df['team'] = team_game_df.team.sample(people_team_df.shape[0], replace=True).to_list()
people_team_df
```

```python
people_team_df.dtypes
```

### people-team-game

```python
people_team_game_df = people_team_df.merge(team_game_df, on=['team'], how='left')
people_team_game_df
```

```python
# write players df to disk

people_team_game_df.to_csv('data/people_team_game.csv', index=False)
```

### datetimes

```python
def pp(start: datetime, end: datetime, n: int):
    # based on https://stackoverflow.com/questions/50559078/generating-random-dates-within-a-given-range-in-pandas
    start_u = int(start.timestamp())
    end_u = int(end.timestamp())
    return pd.DatetimeIndex((np.random.randint(start_u, end_u, n, dtype=np.int64)).view('M8[s]'))
```

### gameplays

```python
num_games = 1000000

gameplays_df = people_team_game_df.sample(num_games, replace=True)
gameplays_df['datetime'] = pp(start=one_year_ago, end=now, n=num_games)
gameplays_df['score'] = (np.exp(np.random.normal(loc=5, scale=1, size=num_games)) + 1).astype(int)
gameplays_df
```
